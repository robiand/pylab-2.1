Setup

```python
from rectangle import Rectangle
from circle import Circle

# create some rectangles
rectangle1 = Rectangle(0, 0, 5, 4) # create a rectangle
rectangle2 = Rectangle(0.4, 0.5, 7, 2) # rectangle with different size
rectangle3 = Rectangle(1, 3, 5, 4) # rectangle with same size as rectangle1
rectangle4 = Rectangle(0, 50, 5, 5)

rectangle1.translate(0.5, 3) # translate rectangle after creation to give x and y pos a non-0 value


# create some circles
circle1 = Circle(0, 0, 7.2) # circle creation
circle2 = Circle(0, 1, 7.5) # create a different circle
circle3 = Circle(0, 0, 1)

circle1.translate(2, 0.2)  # translate a circle
```

Testing

```python
# rectangle testing
print("Rectangle results:")
print(rectangle1 == rectangle2) # compare two rectangles
print(rectangle1 == rectangle3) # compare two rectangles of same size
print(rectangle1.perimeter) # perimeter of a rectangle
print(rectangle4.is_square()) # is rectangle a perfect square?

# circle testing
print("\nCircle results")
print(circle2 == circle1) # compare size of two circles
print(circle2.perimeter) # perimeter value of a circle
print(circle1.is_unit()) # is this a unit circle?
print(circle3.is_unit()) # is this a unit circle?
```
